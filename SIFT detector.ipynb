```python

import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img_path = r"C:\Users\zbook 17 g3\Desktop\New folder (4)\a654cfe4de174885af2f0e67ce78f723.jpeg"
img = cv.imread(img_path)

gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

sift = cv.SIFT_create()
kp, des = sift.detectAndCompute(gray,None)#image &mask

img=cv.drawKeypoints(gray,kp,None,flags=cv.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)


#cv.imshow("image",img)



#cv.waitKey(0)
#cv.destroyAllWindows()
plt.figure(figsize=(20,20))
plt.imshow(img)
plt.axis("off")
plt.title("SIFT detector")
```
